### Knowledge Check
**Question 1:** If you set **$K=3$**, what size will the algorithm recommend for Customer X?  

A) Small

B) Large

C) It is a tie

D) Cannot be determined without normalization


Answer A) Small\
Three smalles neighbours are C(2.24), E(2.83) and B(4.12). All distances are euclidean distance.\
C Size is Large, E size is Small and B size is Small. Therefore size is small\


Calculation:\
Distance = √[(x₁_new - x₁_old)² + (x₂_new - x₂_old)²] \
For Customer X (166, 61): \
• Distance to A: √[(166-158)² + (61-58)²] = √[64 + 9] = 8.54 \
• Distance to B: √[(166-162)² + (61-60)²] = √[16 + 1] = 4.12 \
• Distance to C: √[(166-168)² + (61-62)²] = √[4 + 1] = 2.24 ← Closest \
• Distance to D: √[(166-170)² + (61-64)²] = √[16 + 9] = 5.00 \
• Distance to E: √[(166-164)² + (61-59)²] = √[4 + 4] = 2.83 ← 2nd closest\
The 3 nearest neighbors are:\
Customer C (distance 2.24) - Large\
Customer E (distance 2.83) - Small\
Customer B (distance 4.12) - Small\
Vote count: Small = 2, Large = 1\

The algorithm predicts: SMALL (majority vote)

### Knowledge Check

**Question 2:** If you change the parameter to **$K=1$**, how does the prediction change?  

A) The prediction remains Small; the model is stable.

B) The prediction changes to Large; the model is sensitive to local noise.

C) The prediction changes to Large; the model is underfitting.

D) The prediction becomes a tie; the model needs an odd K.



**Answer: B) The prediction changes to Large; the model is sensitive to local noise.** \
**Explanation:** At K=1, the model only looks at the single closest neighbor.
The closest neighbor to Customer X is Customer C (Large). This shift from the
broader K=3 consensus shows how a low K value can be sensitive to local data points.

### Knowledge Check

**Question 3:** If you store Height in meters (1.66m) but Weight in kg (61kg), which feature will dominate the distance?

The result will not change because KNN is scale-invariant.

B) The "Height" feature will dominate the distance calculation, making Weight irrelevant.

C) The "Weight" feature will dominate the distance calculation, making Height irrelevant.

D) The algorithm will crash because the units do not match.



Answer:
C) The "Weight" feature will dominate the distance calculation, making Height irrelevant.

Explanation: If Height is in meters (1.66 vs 1.68), the squared difference is 0.0004.
If Weight is in kg (61 vs 62), the squared difference is 1.0. Because the scale
of Weight is much larger, it dominates the Euclidean distance formula.

Key concept:
KNN is not scale-invariant → feature scaling (normalization/standardization) is required.

**Question 4:** Output the label for Customer X for each K from 1 to 5. Also, try the code for different combinations of heights and weights.

```python
import pandas as pd
import numpy as np
```

```python
# 1. Prepare the Training Dataset
data = {
    'Customer': ['A', 'B', 'C', 'D', 'E'],
    'Height': [158, 162, 168, 170, 164],
    'Weight': [58, 60, 62, 64, 59],
    'Size': ['Small', 'Small', 'Large', 'Large', 'Small']
}
```

```python
df = pd.DataFrame(data)
```

```python
# 2. Define Customer X
x_height, x_weight = 166, 61
```

```python
# 3. Calculate Euclidean Distances
# Distance = sqrt((x1_new - x1_old)^2 + (x2_new - x2_old)^2)
df['Distance'] = np.sqrt((df['Height'] - x_height)**2 + (df['Weight'] - x_weight)**2)
```

```python
# Sort by distance to find nearest neighbors
df_sorted = df.sort_values(by='Distance').reset_index(drop=True)
```

```python
def predict_size(k):
    # Get the top K neighbors
    neighbors = df_sorted.head(k)
    # Find the most frequent size (mode)
    prediction = neighbors['Size'].mode()[0]
    return prediction, neighbors
```

```python
# --- STUDENT INTERACTIVE SECTION ---
# Change the value of K here (1 to 5)
K_VALUE = [1,2,3,4,5]
#prediction, nearest_neighbors = predict_size(K_VALUE)
```

```python
for i in K_VALUE :
    print(f"Results for K = {i}")
    #print("-" * 30)
    prediction, nearest_neighbors = predict_size(i)
    print(f"Prediction for Customer X: {prediction}")
    #print("-" * 30)
    #print(" " * 30)
    #if i == 3 : 
    #    print("\nNearest Neighbors used:")
    #    print(nearest_neighbors[['Customer', 'Height', 'Weight', 'Distance', 'Size']])
    print()
```

```python
# Experimenting with different combinations (as requested in Q4)
print("\nTesting Custom Combinations:")
test_cases = [(175, 70), (155, 55)]
for h, w in test_cases:
    temp_dist = np.sqrt((df['Height'] - h)**2 + (df['Weight'] - w)**2)
    closest_idx = temp_dist.idxmin()
    print(f"Height {h}cm, Weight {w}kg -> Likely Size: {df.iloc[closest_idx]['Size']} (based on nearest neighbor)")
```
