**Question 2: The Meaning of the Margin**

In SVM terminology, we define the "margin" as the distance between two parallel hyperplanes ($H_1$ and $H_2$) that sit on the edge of the data clusters.



**Which of the following describes the mathematical width of this margin?**

A) $\frac{1}{||{\bf w}||}$

B) ${\bf w}^T{\bf x} + b$

C) $\frac{2}{||{\bf w}||}$

D) $\sum \alpha_i y_i$

#### Answer:

C) $\frac{2}{||{\bf w}||}$

explaination : The margin is defined as the distance between the two hyperplanes H1 and H2 which can be calculated as $\frac{2}{||{\bf w}||}$. this formula comes because the distance from the decision boundary to either hyperplane is $\frac{1}{||{\bf w}||}$ so the total margin width is twice that distance.

### **Part 3: Mathematical Reasoning**

**Question 4: Primal vs. Dual and Kernels**

The lecture explains that we often solve the **Dual** form of the SVM optimization problem rather than the **Primal** form.

**What is a major advantage of solving the SVM in its Dual form?**

A) It removes the need for labels ($y_i$).

B) It allows us to handle non-linear problems using Kernel functions $K({\bf x}_i, {\bf x}_j)$ without explicitly mapping data to higher dimensions.

C) It ensures that the weight vector ${\bf w}$ is always zero.

D) It simplifies the problem into a simple Linear Regression.

#### Answer:

B) It allows us to handle non-linear problems using Kernel functions $K({\bf x}_i, {\bf x}_j)$ without explicitly mapping data to higher dimensions.

explaination : the dual form of the svm optimization problem is expressed in terms of the lagrange multipliers $\alpha_i$ and the inner products of the data points. this allows us to replace the inner product with a kernel function $K({\bf x}_i, {\bf x}_j)$ which can implicitly map the data into a higher-dimensional space without ever computing the coordinates in that space thus enabling us to solve non-linear classification problems efficiently.

### **Part 4: Coding Challenge — SVM Primal Objective**

In this section, you will implement the objective function for a hard-margin SVM in its **Primal** form.

**Formula for Reference:**
Objective: Minimize $\frac{1}{2} ||{\bf w}||^2$ subject to $y_i({\bf w}^T{\bf x}_i + b) \geq 1$.


```python
import numpy as np

# --- 1. Setup Synthetic Data ---
np.random.seed(42)
# 4 points in 2D space
X = np.array([[1, 2], [2, 3], [3, 3], [2, 1]])
y = np.array([1, 1, -1, -1]) # Class labels

# Initial weights and bias
w = np.array([0.5, -0.5])
b = 0.1

# --- 2. The Objective Function ---

def svm_primal_objective(w):
    """
    Computes the term we want to minimize in the primal SVM: 0.5 * ||w||^2
    """
    # -----------------------------------------------------
    # TODO: Calculate 0.5 * squared norm of w
    # Hint: Use np.dot or np.linalg.norm
    objective = 0.5 * np.dot(w, w)  
    # -----------------------------------------------------
    return objective

def check_constraints(X, y, w, b):
    """
    Checks if the points satisfy the SVM constraint: y_i(w.x + b) >= 1
    """
    # -----------------------------------------------------
    # TODO: Calculate the scores for all points
    # score = y * (Xw + b)
    scores = y * (X.dot(w) + b)
    
    # Check which points satisfy the margin constraint (score >= 1)
    satisfied = scores >= 1
    # -----------------------------------------------------
    return satisfied, scores
```

```python
# --- 3. Execution & Verification ---

try:
    obj = svm_primal_objective(w)
    satisfied_mask, all_scores = check_constraints(X, y, w, b)

    print(f"Current Objective Value: {obj}")
    print(f"Constraint Satisfaction: {satisfied_mask}")
    print(f"Scores (should be >= 1 for success): {all_scores}")

    # Integrity Check
    if obj is not None and all_scores is not None:
        print(f"\n1. Objective Math Integrity: {'✅ Pass' if np.isclose(obj, 0.25) else '❌ Fail'}")
        print(f"2. Constraint Logic Sync:    {'✅ Pass' if len(all_scores) == len(y) else '❌ Fail'}")
except:
    print("Function incomplete!")
```
